```python
cases = pyger.read_cases('cases_for_plot_current_with_fwdblkhd_5ccd_2015001.csv')
#data, coolstats, hotstats = pickle.load(open('../pygeroutput/pyger_single_msid_2015001_pftank2t_93_4ccd_dwell2.pkl','r'))
```

Generate the composite dataset.

```python
post = pyger.PostPyger(cases, '../pygeroutput/')
```

Add PLINE Limit Info

```python
dwell_156_162 = 60*(9*60 + 20)
dwell_156_166 = 60*(7*60 + 50)
dwell_156_170 = 60*(4*60 + 20)
post.max_hot_time.max_time[156] = dwell_156_162
post.max_hot_time.max_time[157] = dwell_156_162
post.max_hot_time.max_time[158] = dwell_156_162
post.max_hot_time.max_time[159] = dwell_156_162
post.max_hot_time.max_time[160] = dwell_156_162
post.max_hot_time.max_time[161] = dwell_156_162
post.max_hot_time.max_time[162] = dwell_156_162
post.max_hot_time.max_time[163] = dwell_156_166
post.max_hot_time.max_time[164] = dwell_156_166
post.max_hot_time.max_time[165] = dwell_156_166
post.max_hot_time.max_time[166] = dwell_156_166
post.max_hot_time.max_time[167] = dwell_156_170
post.max_hot_time.max_time[168] = dwell_156_170
post.max_hot_time.max_time[169] = dwell_156_170
post.max_hot_time.max_time[170] = dwell_156_170

for n in range(156, 171, 1):
    post.max_hot_time.limiting_msid[n] = 'PLINE'

```

Use a maximum dwell time in place of Nans.

```python
indnan = np.isnan(post.max_hot_time.max_time)
maxtimes = post.max_hot_time.max_time
maxtimes[indnan] = 600000
```

```python
msids = ['1PDEAAT','TCYLAFT6', 'PFTANK2T',  'AACCCDPT', '4RT700T', '1DPAMZT', 'PLINE']
colors = ['#E69F00', '#009E73','#FF3900',  '#0072B2', '#D55E00', '#CC79A7', '#004F25']

fig = plt.figure(figsize=[14, 9])
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(post.pitch_set, maxtimes, color=[0.4, 0.4, 0.4], linewidth=3, label='Max Dwell Capability')
for msid, color in zip(msids, colors):
    ind = np.array(post.max_hot_time.limiting_msid == msid)
    ax.plot(np.array(post.pitch_set)[ind], maxtimes[ind], 'o', color=color, label=msid, markersize=10)

ax.set_yticks(range(0, 550000, 50000))
ax.set_yticklabels(range(0, 550, 50), fontsize=20)
ax.set_ylabel('Dwell Time (Kiloseconds)', fontsize=24)
ax.set_ylim(0, 500000)

ax.set_xticks(range(40, 180, 10))
ax.set_xticklabels(range(40, 180, 10), fontsize=20)
ax.set_xlabel('Dwell Pitch', fontsize=24)
ax.set_xlim(40, 170)

ax.legend(loc='best', fontsize=16)
ax.grid(True)
ax.set_title('2014:001 5CCD Max Dwell Composite', fontsize=28)
fig.savefig('Composite_2014001_5ccd_max_dwell-generated_2014080.png')
```
